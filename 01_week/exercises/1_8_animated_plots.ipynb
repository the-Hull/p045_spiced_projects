Now convert the table into long format. First, we have to move the row index into a column:


Second, we use the convenient melt function for the conversion:

```python

gaplist = [
    fert,
    life,
    pop
]


# rename all columns to integer values
gaplist = map(
    lambda x: x.rename(mapper = lambda y: int(y), axis = 'columns'),
    gaplist
)



gaplist = list(
    map(
        lambda x: x.reset_index(),
        gaplist
    )
)




gaplist = list(
    map(
        lambda x: x.rename(columns = {x.columns.values[0]: 'country'}),
        gaplist
    )
)

# [print(x.head()) for x in gaplist]

gaplist_melt = list(
    map(
        lambda x, y: x.melt(id_vars='country', var_name='year', value_name=y),
        gaplist,
        ['fertility_rate',
        'life_expectancy',
        'population_size']
    )
)

[print(x.head()) for x in gaplist_melt]
```

## Step 11
To create an animated scatterplot showing the correlation between life expectancy and fertility, create one plot for each year between 1960 and 2015 (before that, the data contains too many gaps).

Write a for loop that goes through each year and produces a scatterplot.

Save each scatterplot to a separate file containing the year in the filename, e.g. lifeexp_1999.png. After saving the plot call plt.close() to remove the plot from the notebook.

```python
# check to get good values for plot lims

gap_join.describe()
```

```python

def save_year_plots(df, yrs, path_dir):
    
    for y in yrs:
        plt.figure(figsize = (6,6))
        df_subset = df.loc[df['year'] == y]
        sns.scatterplot(
            x='life_expectancy',
            y='fertility_rate',
            size='population_size',
            data=df_subset,
            alpha=0.6
            )
        path_file = path_dir + f'fig_{y}.png'

        plt.title(f'Year: {y}')
        plt.axis((0, 110, 0, 10))
        plt.legend([],[], frameon=False)

        plt.savefig(path_file, facecolor = 'white')
        plt.close()

    return "Finished figs at " + path_dir
```

```python

years = range(1960, 2015+1)
# [print(y) for y in years]
save_year_plots(
    df = gap_join,
    yrs = years,
    path_dir = "1_8_imgs/"
    )
```

```python
# grab files and years from files
import os 
import re
```

```python
files = os.listdir("1_8_imgs")

years_from_files = []

for f in files:
    # use regular expression to grab years
    years_from_files.append(re.search('(\d{4})', f).group())


years_from_files = sorted([int(x) for x in years_from_files])
```

```python
import imageio

images = []

for i in years_from_files:
    filename = '1_8_imgs/fig_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('output.gif', images, fps=20)
```
