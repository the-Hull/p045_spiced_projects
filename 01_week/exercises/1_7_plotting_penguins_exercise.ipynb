### 1. Do heavier penguins have longer flippers?

#### Conceptually
- Which columns do we need?
- Which plots do we use?

```python
labsize = 20
plt.figure(figsize=(10,6))
plt.scatter(x = df['Body Mass (g)'], y = df['Flipper Length (mm)'])
plt.xlabel('Body Mass (g)', fontsize = labsize)
plt.ylabel('Flipper Length (mm)', fontsize = labsize)
```

- Which plots do we use?

```python
colmap = pd.DataFrame(
    {
        'Species' : df['Species'].unique(),
        'color' : ['black', 'blue', 'green']
        
    }
)

df = pd.merge(df, colmap,  on = 'Species')
```

```python
plt.scatter(x = df['Body Mass (g)'], y = df['Flipper Length (mm)'], color = df['color'])
```

Use df subsets to plot groups, set `label = 'Whatever'` to capture grouping in legend

```python
coldic = dict(zip(colmap.Species, colmap.color))
plt.scatter(x = df['Body Mass (g)'], y = df['Flipper Length (mm)'], c = df['Species'].map(coldic))

```

### 2. How similar are penguins in their body size?

```python
### Look at the distribution 
plt.hist(df['Body Mass (g)'])
plt.title('Body mass distribution')
plt.xlabel('Body Mass (g)')
plt.ylabel('Frequency')
```

### Is there a difference between males and females?

```python
plt.hist(df.loc[df['Sex'] == 'FEMALE', 'Body Mass (g)'], label = "F", alpha = 0.6)
plt.hist(df.loc[df['Sex'] == 'MALE', 'Body Mass (g)'], label = "M", alpha = 0.6)
plt.legend()
plt.title('BM Distribution')
plt.xlabel('Body Mass (g)')
plt.ylabel('Count')
```
