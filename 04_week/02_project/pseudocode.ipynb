# Developing functions with pseudo code

## What?

Plain-language instructions to reach a goal --> **Algorithm**

## Why?

Thinking purely in code is challenging. **Conceptualizing** an algorithm is not!

## How?

- Define Goal
- Define inputs & outputs
- Write out major steps
    - in a text file
    - as comments in code (**!**) ###
    - think about inputs and outputs for each step
    - bonus: you can start developing _anywhere_
- Write out minor steps for a given major step
- Write (dummy) code
- Develop --> Evaluate --> Test
